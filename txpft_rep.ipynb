```python
import pandas as pd
import os
import glob
import shutil
```

```python
PFT_ANOT_PATH = "/home/jeevan/Desktop/pft annotated"
```

```python
FILE_PARTS = ["app-version-code", "subject-ID", "file-class", "file-ID", "file-comment", "file-format", "file-name", "file-path"]

def process_file_info(file_path: str, SEP="-") -> dict:

    file_name: str = os.path.basename(file_path)
    file_name_parts = file_name.replace(".", SEP).split(sep=SEP)
    file_name_parts += [file_name, file_path]

    file_info_dict = { p:file_name_parts[i] for i, p in enumerate(FILE_PARTS) }

    return file_info_dict
```

```python
all_data_files = glob.glob(f"{PFT_ANOT_PATH}/pft_tsv/*.tsv")

all_data_files_DF = pd.DataFrame([process_file_info(f) for f in all_data_files])
all_data_files_DF
```

```python
def read_pft_data(row):
    file_path = row[f'file-path']

    df = pd.read_csv(file_path, sep=",")

    nfile_path = file_path.replace(".csv", ".tsv")

    df.to_csv(nfile_path, index=False, sep='\t')

# Test
read_pft_data(all_data_files_DF.iloc[0])
```
