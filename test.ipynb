```python
import string
import random
import uuid
import pandas as pd
```

```python
def random_string(N=7):
    s = ''.join(random.choices(string.ascii_lowercase, k=N))
    c = uuid.uuid4().hex[:8]
    return "_".join([s, c])

random_string()
```

```python
S = 10
SUBS = [random_string() for _ in range(S)]
SUBS
```

```python
APP_CODE = "pnoistor_feb2023"
CLASS = "CLASS"
COMNT = "comnt"
TYPE = "TYPE"

def get_file_name(sid):
    hs = uuid.uuid4().hex[:4]
    s = "-".join([APP_CODE, sid, CLASS, hs, COMNT])
    return s + f".{TYPE}"

get_file_name(SUBS[0])
```

```python
file_names_df = pd.DataFrame([fn.split("-") for fn in file_names])
file_names_df.columns = ["app_code", "sid", "class", "hs", "ext"]
file_names_df
```

```python
import hashlib
import os

```

```python
def create_anotes(filename, bb=False, n=1):
    VBA_labels = [  
        "aa", "ee", "uu", "oo", 
        "ii", "xx",  "ii", "xx", "ii", "xx", "ii", "xx", "ii", "xx", 
        "aa", "ee", "uu", "oo"
    ]

    VBA_labels_df = pd.DataFrame(VBA_labels, columns=["label"])

    VBA_labels_df["start"] = list(range(0, len(VBA_labels_df) * 2, 2))
    VBA_labels_df["start"] += 2.3
    VBA_labels_df["end"] = VBA_labels_df["start"] + 2

    def inc_df(n):
        _df = VBA_labels_df.copy()

        _df["start"] += n * 50
        _df["end"] += n * 50
        _bb = [{
                "start": _df["start"][0],
                "end": _df["end"][len(_df) - 1],
                "label": f"bb{n + 1}",
        }]
        _bb_df =  pd.DataFrame(_bb)

        _df = pd.concat([_df, _bb_df])

        return _df


    VBA_labels_full_df = pd.concat([inc_df(n)  for n in range(0, 4)])

    if bb:

        _bb = [{
                "start": 2.3,
                "end": 38.3,
                "label": f"bb{n}",
        }]

        VBA_labels_full_df =  pd.DataFrame(_bb)

    VBA_labels_full_df.to_csv(filename, columns=["start", "end", "label"], sep="\t", index=False)




    return VBA_labels_full_df

create_anotes("test.txt", bb=False, n=1)
```

```python
def file_hash(seed):
    m = hashlib.md5()
    m.update(seed.encode('utf-8'))
    new_uuid = uuid.UUID(m.hexdigest())
    return new_uuid.hex[:4]

DATA = "DATA"
if (not os.path.exists(DATA)):
    os.mkdir(DATA)

def make_sub():
    _file_name_df = file_names_df.copy()

    sid = random_string()
    _file_name_df["sid"] = sid
    _file_name_df["hs"] = [file_hash(r["sid"] + r["class"]) for i, r in _file_name_df.iterrows() ]

    os.mkdir(f"{DATA}/{sid}")

    return _file_name_df

sub_files = make_sub()

for i, fn in sub_files.iterrows():
    fname = "-".join(fn.to_dict().values())

    sid =  fn["sid"]
    fpath = f"{DATA}/{sid}/{fname}"
    open(fpath, 'a').close()
    
    if (".txt" in fname):
        if ("VBA" in fname):
            create_anotes(fpath)
        if ("LBA" in fname):
            if("LU" in fname):
                create_anotes(fpath, bb=True, n=1)

            if("RU" in fname):
                create_anotes(fpath, bb=True, n=2)

            if("LL" in fname):
                create_anotes(fpath, bb=True, n=3)

            if("RL" in fname):
                create_anotes(fpath, bb=True, n=4)
            


    # break
```
